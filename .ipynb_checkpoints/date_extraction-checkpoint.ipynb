```python
import os
import pytesseract
import pdfplumber
from pdf2image import convert_from_path
import numpy as np
import nltk
from unidecode import unidecode
import itertools


def get_text_from_pdf(path):
    pages = convert_from_path(path, 350)
    imgBlobs = []
    extracted_text = [pytesseract.image_to_string(img, lang='por') for img in pages]
    return (extracted_text)


def read_file(pdf_path):
    pdf = pdfplumber.open(pdf_path)
    text = []
    for page in pdf.pages:
        text.append(page.extract_text())
    if None in text:
        print("Handling file as image...")
        text = get_text_from_pdf(pdf_path)
    text = ' (NEWPAGE) '.join(text)
    return text


def clean_text(text):
    stop_words = nltk.corpus.stopwords.words("portuguese")
    text_by_words = text.replace('\n', ' ').split(' ')
    formated_text = [unidecode(word.lower()) for word in text_by_words if (len(word) >= 1) & (word.lower() not in stop_words)]
    return formated_text


def get_prazos(text):
    sentences = []
    for idx, i in enumerate(text):
        if 'vigencia' in i:
            sentences.append(' '.join(text[idx-5: idx+15]))
    return sentences


def read_all_documents(pdf_path):
    prazos = []
    for file in sorted(os.listdir(pdf_path))[:4]:
        print('Reading File {}'.format(file))
        text = read_file('{}/{}'.format(pdf_path, file))
        print('Cleaning Texts')
        text = clean_text(text)
        print('Getting Prazos')
        sentences = get_prazos(text)
        prazos.append(sentences)
    return prazos

def write_prazos(prazos):
    with open('output.txt', 'w') as f:
        for i in range(len(prazos)):
            f.write('FILE: {}\n'.format(i))
            for _string in prazos[i]:
                #f.seek(0)
                f.write(str(_string) + '\n')
            f.write('\n')
```

```python
prazos = read_all_documents('./documents/contratante')
write_prazos(prazos)
```
